```python
import ast

def parse_crew(crew_str):
    if pd.isna(crew_str) or crew_str == "":
        return []  
    try:
        return ast.literal_eval(crew_str)
    except Exception as e:
        print("Error parsing crew:", e)
        return []  

data['crew'] = data['crew'].apply(parse_crew)
```
